# Introdução ao Matplotlib

## Gráficos no Python

```python
import matplotlib.pyplot as plt #".pyplot" é conjunto de métodos que serão utilizados
import numpy as np  #funções matemáticas
```

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], 'ro') #apenas pontos
plt.show()
```

```python
x = np.arange(0, 10, 0.1) #lista de 0 a 10 em 0.1
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 10)) #plotar mais de um gráfico
fig1.suptitle('Minha primeira figura!', size=30) #Título para a figura toda
#plt.style.use('seaborn')

f1_axes[0, 0].plot(np.sin(x), label='sen(x)', color ='darkturquoise') #passa os valores de x na função seno
f1_axes[0, 0].plot(np.cos(x), label='cos(x)', color = 'pink') #plota por cima no mesmo gráfico
f1_axes[0, 0].set_title('Caixa 1') #título que pode ser definido para cada um dos subplots
f1_axes[0, 0].set_xlabel('Caixa 1 - Label X') #rótulo do eixo x
f1_axes[0, 0].set_ylabel('Caixa 1 - Label Y') #rótulo do eixo y
f1_axes[0, 0].legend() #legenda para o gráfico. Ele plota no gráfico a legenda baseada no label de cada plot
f1_axes[0, 1].plot(np.cos(x))
f1_axes[0, 1].set_title('Caixa 2')
f1_axes[0, 1].set_xlabel('Caixa 2 - Label X')
f1_axes[0, 1].set_ylabel('Caixa 2 - Label Y')

f1_axes[1, 0].plot(np.tan(x))
f1_axes[1, 0].set_title('Caixa 3')
f1_axes[1, 0].set_xlabel('Caixa 3 - Label X')
f1_axes[1, 0].set_ylabel('Caixa 3 - Label Y')

f1_axes[1, 1].plot(np.sin(x))
f1_axes[1, 1].set_title('Caixa 4')
f1_axes[1, 1].set_xlabel('Caixa 4 - Label X')
f1_axes[1, 1].set_ylabel('Caixa 4 - Label Y')
```
