```python
import pandas as pd
```

```python
caminho = "D:\ARQUIVOS\MICRODADOS_ENEM_2019.csv"    #path para importação de arquivos
colunas = ['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
'NU_NOTA_MT', 'NU_NOTA_REDACAO']    #colunas selecionadas para importação
microdadosEnem = pd.read_csv(caminho, sep=';', encoding='ISO-8859-1', usecols=colunas) #microdados armazenado como data frame
```

```python
mediaEnemPorMunicipio = microdadosEnem.groupby(['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA'], as_index=False).mean()
```
